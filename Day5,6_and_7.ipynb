```
math=np.array(students.iloc[:,5].values)
reading=np.array(students.iloc[:,6].values)
print('math',math[:5] )
print('reading', reading[:5])
```

```
math = math.reshape( -1, 1)
print('math',math[:5])
```

```
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(math,reading)
model.score(math,reading)
```

```
import matplotlib.pyplot as plt
plt.scatter(math,reading)
plt.plot(np.array(students.iloc[:,5].values),model.predict(math),'r')
```
